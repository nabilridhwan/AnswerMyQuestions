```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
sns.set_theme(style="darkgrid")
```

```python
covid_world_dataset = pd.read_csv("./datasets/worldometer_data.csv")

top_covid_countries = covid_world_dataset.sort_values(by="TotalCases", ascending=False).head(10)

plt.title("Countries with the most number of COVID-19 cases")
sns.barplot(x="Country/Region", y="TotalCases", data=top_covid_countries)
plt.xlabel("Country")
plt.ylabel("Total Cases (In Millions)")
plt.xticks(rotation=45);
```

```python
covid_world_dataset = pd.read_csv("./datasets/worldometer_data.csv")

top_covid_countries = covid_world_dataset.sort_values(by="Tot Cases/1M pop", ascending=False).head(10)

plt.title("Countries with the most number of COVID-19 cases per 1M population")
sns.barplot(x="Country/Region", y="Tot Cases/1M pop", data=top_covid_countries)
plt.xlabel("Country")
plt.ylabel("Total Cases (In Millions)")
plt.xticks(rotation=45);
```
